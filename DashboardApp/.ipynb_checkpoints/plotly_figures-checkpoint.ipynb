```python
for col in df.columns:
    print(col)
    display(df[col].unique())
```

```python
df.gender.replace({0: 'Male', 1:'Female'}, inplace=True)
```

```python
df.household_income.replace({0:'Missing', 
                             1:'Below $5K', 
                             2:'Below $10K', 
                             3:'Below $15K', 
                             4:'Below $20K', 
                             5:'Below $25K', 
                             6:'Below $35K', 
                             7:'Below $45K', 
                             8:'Below $55K', 
                             9:'Below $65K', 
                             10:'Below $75K',
                             11:'$75K+',
                             12:'Over $20K', 
                             13:'Under $20K', 
                             14:'Below $100K', 
                             15:'$100K+'}, inplace=True)
```

```python
df.current_smoker.replace({0:'Missing', 
                           1:'Every Day', 
                           2:'Some Days', 
                           3:'Never'}, inplace=True)
```

```python
df.cocaine_number_uses.replace({0:'Missing', 
                                1:'Once', 
                                2:'2-5', 
                                3:'6-19', 
                                4:'20-49', 
                                5:'50-99', 
                                6:'110+'}, inplace=True)
```

```python
df.meth_number_uses.replace({0:'Missing', 
                                1:'Once', 
                                2:'2-5', 
                                3:'6-19', 
                                4:'20-49', 
                                5:'50-99', 
                                6:'110+'}, inplace=True)
```

```python
yes_no_cols = list(df.columns)
temp_list = [0, 1]
for col in df.columns:
    x = list(sorted(df[col].unique()))
    if x != temp_list:
        yes_no_cols.remove(col)
```

```python
yes_no_cols
```

```python
for col in yes_no_cols:
    df[col].replace({0:'No', 1:'Yes'}, inplace=True)
```

```python
for col in df.columns:
    print(col)
    display(df[col].unique())
```

```python
list(df.columns)
```

```python
for col in df.columns:
    new_name = col.replace('_', ' ').title()
    df.rename(columns={col: new_name}, inplace=True)
```

```python
list(df.columns)
```

```python
upper_case = ['Bmi',
 'Hdl',
 'Ldl',
 'Alt',
 'Ast',
 'Alp',
 'Bun',
 'Co2',
 'Ggt',
 'Lhd',
 'Mcv',
 'Mch',
 'Mchc',
 'Rdw',
 'Mpv']

for col in upper_case:
    new_name = col.upper()
    df.rename(columns={col: new_name}, inplace=True)
```

```python
df.rename(columns={'Triglyceride': 'Triglycerides',
                   'White Bcc': 'White BCC', 
                   'Rbc Count': 'RBC Count'}, inplace=True)
```
